```python
from rdkit.Chem import AllChem as Chem
Chem.MolToSmiles(Chem.MolFromSLN('N[1](CH2CH2CH2C@1=O)CH3'))
'CN1CCCC1=O'
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/physical property predictions/entropy/entropy_data/df_smiles.csv")
```

```python
df
```

```python
df1 = df.drop_duplicates()
```

```python
df1
```

```python
name = df1['isomeric_smiles'].tolist()
molecules =[]
for i in name:
  molecules.append(Chem.MolFromSmiles(i))
if __name__ == "__main__":
    freeze_support()
    # Create Calculator
    calc = Calculator(descriptors)
df2= pd.DataFrame(data=calc.pandas(molecules))
df2
```

```python
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
extracted_col_name= df1["name"]
extracted_col_hfg= df1["hfg"]
df2.insert(0, "name", extracted_col_name)
df2 = df2.join(extracted_col_hfg)
df2
```
